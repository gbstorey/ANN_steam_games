```python
import pandas as pd
import requests as r
import time
import os
from dotenv import load_dotenv
load_dotenv()
```

```python
CLIENT_ID = os.environ['CLIENT_ID']
CLIENT_SECRET = os.environ['CLIENT_SECRET']
grant_type = "client_credentials"
payload = {"client_id": CLIENT_ID,
           "client_secret": CLIENT_SECRET,
           "grant_type": grant_type}
res = r.post('https://id.twitch.tv/oauth2/token', params=payload).json()
```

```python
IGDB_URL = "https://api.igdb.com/v4/games"
AUTHORIZATION = "Bearer " + res['access_token']
headers = {
    "Client-ID": CLIENT_ID,
    "Authorization": AUTHORIZATION,
}
```

```python
df = pd.read_csv("steamgameinfo.csv")
names = df.name.tolist()
df.head()
```

```python
ratings = []
for index, name in enumerate(names):
    raw_data = f'fields *; where name = "{name}";'.encode('utf-8')
    res = r.post(IGDB_URL, data=raw_data, headers=headers).json()
    try:
        rating = res[0].get("rating")
    except IndexError:
        print(f"{name} not found")
        rating = None
    ratings.append(rating)
    time.sleep(1)
    print(f"{name} Complete!")
df.rating = df.Series(ratings)
df.head()
```

```python
df["ratings"] = ratings
df.to_csv("with_ratings.csv")
```
